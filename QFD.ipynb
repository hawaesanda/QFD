```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data for the QFD matrix
customer_reqs = [
    "Login mudah", "Manajemen data", "Absensi & lembur",
    "Dokumen kerja", "Pembayaran fleksibel",
    "Akses mobile & web", "Laporan rapi", "UI modern"
]

technical_reqs = [
    "Autentikasi", "Manajemen Karyawan", "Absensi",
    "Lembur", "Dokumen", "Payment Gateway",
    "Laporan", "UI/UX"
]

# Matrix values (same order as described earlier)
qfd_matrix = np.array([
    [9, 0, 0, 0, 0, 0, 0, 0],
    [0, 9, 0, 0, 3, 0, 3, 0],
    [0, 0, 9, 9, 0, 0, 6, 0],
    [0, 0, 0, 0, 9, 0, 3, 0],
    [0, 0, 0, 0, 0, 9, 6, 0],
    [0, 0, 0, 0, 0, 0, 0, 9],
    [0, 0, 3, 3, 3, 0, 9, 0],
    [0, 0, 0, 0, 0, 0, 0, 9]
])

# Create the DataFrame
df = pd.DataFrame(qfd_matrix, index=customer_reqs, columns=technical_reqs)

# Plotting the House of Quality matrix
fig, ax = plt.subplots(figsize=(12, 8))
cax = ax.matshow(df, cmap="Blues")

# Show all ticks and label them
ax.set_xticks(np.arange(len(technical_reqs)))
ax.set_yticks(np.arange(len(customer_reqs)))
ax.set_xticklabels(technical_reqs, rotation=45, ha="left")
ax.set_yticklabels(customer_reqs)

# Show values in each cell
for (i, j), val in np.ndenumerate(qfd_matrix):
    if val > 0:
        ax.text(j, i, f'{val}', ha='center', va='center', color='black', fontsize=10)

plt.title("House of Quality - HRIS Project", pad=20, fontsize=14)
fig.tight_layout()
plt.colorbar(cax, label='Relevansi')
plt.grid(False)
plt.show()
```
